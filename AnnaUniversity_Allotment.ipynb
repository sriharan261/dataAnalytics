#Join the all the DataFrames

```python
res=pd.merge(df,r1_con,on=['COLLEGE_CODE','BRANCH_CODE'],how='left')
```

```python
res1=pd.merge(res,r2_con,on=['COLLEGE_CODE','BRANCH_CODE'],how='left')
```

```python
res2=pd.merge(res1,r3_con,on=['COLLEGE_CODE','BRANCH_CODE'],how='left')
```

#Filling NAN with Zeros

```python
result=res3.fillna(0)
```

```python
result
```

```python
result["ROUND_2"]=result["ROUND_1"]+result["ROUND_2"]
result["ROUND_3"]=result["ROUND_2"]+result["ROUND_3"]
result["ROUND_4"]=result["ROUND_3"]+result["ROUND_4"]
```

```python
result.head(50)
```

```python
result["ROUND_1"]=round((result["ROUND_1"]/result["TOTAL"])*100,2)
```

```python
result["ROUND_2"]=round((result["ROUND_2"]/result["TOTAL"])*100,2)
```

```python
result["ROUND_3"]=round((result["ROUND_3"]/result["TOTAL"])*100,2)
```

```python
result["ROUND_4"]=round((result["ROUND_4"]/result["TOTAL"])*100,2)
```
